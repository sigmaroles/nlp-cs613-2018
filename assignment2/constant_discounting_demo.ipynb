```python
%load_ext autoreload
%autoreload 2
from ngram_util import NGramModel
import matplotlib.pyplot as plt
import matplotlib
```

```python
with open('corpus_ALL.txt', 'r', encoding='utf-8') as fh:
    corpus = fh.read()
corpus = corpus.split(' ')
    
n = 2

matplotlib.rcParams['figure.figsize'] = (24,10)
```

```python
model = NGramModel(n,corpus,smoothing='goodturing')
model.buildModels()
```

```python
d_vals = model.get_d_list()
plt.plot(d_vals, lw=0, markersize=2, marker='o', color='orange')
plt.ylim(-1,5)
plt.xlabel('n grams')
plt.ylabel('proposed \'d\' value')
plt.savefig('d_plot.svg', format='svg')
plt.show()
```

#### Any of the "straight lines" from the plot above could serve as a value of d. But there are too many such lines, not sure about relative performance (if GT smoothing is replaced with that value as constant discounting)

```python
sentences = ['see it pop down a large rabbit hole under the hedge </s>', 
             'either the locks were too large or the key was too small but at any rate it would not open any of them',
             'oh my poor little feet i wonder who will put on your shoes and stockings for you now dears',
             'this is becoming too confusing',
             'she thought this is going to be quite easy',
             'if only we knew how to proceed'
            ]
for sent in sentences:
    print ("probability(natural logarithm) = {}, perplexity = {}".format(model.get_probability(sent), model.get_perplexity(sent)))
    
```
