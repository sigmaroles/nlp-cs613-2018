```python
%load_ext autoreload
%autoreload 2
from ngram_util import NGramModel
import matplotlib.pyplot as plt
import matplotlib
```

```python
with open('corpus_ALL.txt', 'r', encoding='utf-8') as fh:
    corpus = fh.read()
corpus = corpus.split(' ')
    
n = 2

matplotlib.rcParams['figure.figsize'] = (24,10)
```

```python
model = NGramModel(n,corpus,smoothing='goodturing')
model.buildModels()
```

```python
d_vals = model.get_d_list()
plt.plot(d_vals, lw=0, markersize=2, marker='o', color='orange')
plt.ylim(-1,5)
plt.xlabel('n grams')
plt.ylabel('proposed \'d\' value')
plt.savefig('d_plot.svg', format='svg')
plt.show()
```
